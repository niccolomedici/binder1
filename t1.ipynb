```python
prob = 0.5  # probability for a girl
```

```python
births = 8
```

```python
girls = [1000 * prob**n for n in range(1,births+1)]
```

```python
boys  = [1000 * prob**(n-1) * (1-prob) for n in range(1,births+1)]
```

```python
from matplotlib.pyplot import *

labels = [str(n+1) + ' הדיל' for n in range(0, len(boys))]
width = 0.93  # width of bars

barh(labels, girls, width, label='תונב', color='pink')
barh(labels, boys, width, label='םינב', color='blue', left=girls)
xlabel('תודלוי')
xticks([])
legend()
show()
```
