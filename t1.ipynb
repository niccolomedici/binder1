```python
prob = 0.4  # probability for a boy
```

```python
births = 9
```

```python
girls = [1000 * (1-prob)**n for n in range(1,births+1)]
```

```python
boys = [1000 * ((1-prob)**(n-1)) * prob for n in range(1,births+1)]
```

```python
from matplotlib.pyplot import *

labels = [ str(n+1) + ' הדיל' for n in range(0, len(boys))]
width = 0.93  # width of bars

barh(labels, girls, width, label='תונב', color='pink')
barh(labels, boys, width, label='םינב', color='blue', left=girls)
xlabel('תודלוי')
xticks([])
legend()
show()
```
